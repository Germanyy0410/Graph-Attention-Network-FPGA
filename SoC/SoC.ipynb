```python
import numpy as np
weight = "weight.txt"
node_info = "node_info.txt"
h_data = "h_data.txt"

def read_data_from_file(file_path):
    raw_input = []
    with open(file_path, "r") as file:
        for line in file:
            raw_input.append(line.strip())
    return raw_input  


weight_bin = read_data_from_file(h_data)
node_info_bin = read_data_from_file(node_info)
h_data_bin = read_data_from_file(h_data)

```

```python
def input_converter(input_file, signed=True):
  decimal_list = []
  with open(input_file, 'r') as infile:
    for line in infile:
      binary_str = line.strip()
      if not binary_str:
        continue
      try:
        bit_length = len(binary_str)
        if signed:
          decimal_value = int(binary_str, 2) - (1 << bit_length) if binary_str[0] == '1' else int(binary_str, 2)
        else:
          decimal_value = int(binary_str, 2)
        decimal_list.append(decimal_value)
      except ValueError:
        print(f"Skipping invalid binary line: {binary_str}")
        continue
  return decimal_list

input_weight = input_converter(weight, False)
print(input_weight)
input_node_info = input_converter(node_info, False)
print(input_node_info)
input_h_data = input_converter(h_data, False)
print(input_h_data[:11])
```
